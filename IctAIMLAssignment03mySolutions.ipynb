Data preprocessing

clean numeric columns

```python
df["Engine"] = df["Engine"].astype(str).str.replace(" cc", "", regex=False)
df["Engine"] = pd.to_numeric(df["Engine"], errors="coerce")
```

```python
df["Max Power"] = df["Max Power"].astype(str).str.extract(r'(\d+\.?\d*)')
df["Max Torque"] = df["Max Torque"].astype(str).str.extract(r'(\d+\.?\d*)')
df["Max Power"] = pd.to_numeric(df["Max Power"], errors="coerce")
df["Max Torque"] = pd.to_numeric(df["Max Torque"], errors="coerce")
```

```python
df.isnull().sum()
```

```python
# Important numeric columns
num_cols = ["Year","Kilometer","Engine","Max Power","Max Torque",
            "Length","Width","Height","Seating Capacity","Fuel Tank Capacity"]

# Important categorical columns
cat_cols_small = ["Fuel Type","Transmission","Seller Type"]

# Create new dataframe with target
df_small = df[num_cols + cat_cols_small + ["Price"]]

# Fill missing categorical values with mode
for col in cat_cols_small:
    df_small[col] = df_small[col].fillna(df_small[col].mode()[0])

# One-hot encode categorical columns
df_encoded = pd.get_dummies(df_small, drop_first=True)

# Check first few rows
df_encoded.head()
```

Feature scaling

```python
from sklearn.preprocessing import StandardScaler

# Separate features and target
X = df_encoded.drop("Price", axis=1)
y = df_encoded["Price"]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Check shapes
X_scaled.shape, y.shape
```

Model Development

Train Linear Regression model

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.25, random_state=42
)

# Check shapes
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression

# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Check first 5 coefficients and intercept
print("First 5 coefficients:", model.coef_[:5])
print("Intercept:", model.intercept_)
```

```python
# Predict on test set
y_pred = model.predict(X_test)

# Compare first 10 predicted vs actual prices
results = pd.DataFrame({
    "Actual Price": y_test.values,
    "Predicted Price": y_pred.astype(int)
})

results.head(10)
```

Evalute model

```python
from sklearn.metrics import r2_score, mean_absolute_error

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"R² Score: {r2:.4f}")
print(f"Mean Absolute Error: ₹{int(mae):,}")
```

predicting price for new cars

```python
# Example new cars
new_cars = pd.DataFrame({
    "Year": [2018, 2015],
    "Kilometer": [50000, 80000],
    "Engine": [1197, 1498],
    "Max Power": [82, 110],
    "Max Torque": [113, 140],
    "Length": [3995, 4200],
    "Width": [1680, 1750],
    "Height": [1505, 1550],
    "Seating Capacity": [5, 5],
    "Fuel Tank Capacity": [37, 45],
    "Fuel Type": ["Petrol", "Diesel"],
    "Transmission": ["Manual", "Automatic"],
    "Seller Type": ["Individual", "Dealer"]
})

# One-hot encode categorical columns (same as training)
new_cars_encoded = pd.get_dummies(new_cars)
# Add missing columns with 0
for col in X.columns:
    if col not in new_cars_encoded.columns:
        new_cars_encoded[col] = 0
# Reorder columns to match training
new_cars_encoded = new_cars_encoded[X.columns]

# Scale features
new_cars_scaled = scaler.transform(new_cars_encoded)

# Predict
pred_prices = model.predict(new_cars_scaled)
for i, price in enumerate(pred_prices):
    print(f"Predicted Price for Car {i+1}: ₹{int(price):,}")
```
